# Matplotlib
https://matplotlib.org/

#### matplotlib 시각화 예제 갤러리
https://matplotlib.org/stable/gallery/index.html

- 파이썬의 자료(DataFrame, Series 등..)을  차트(chart) 나 플롯(plot) 으로 시각화(visualization) 하는 모듈이다.
- Matplotlib는 다음과 같은 정형화된 차트나 플롯 이외에도 저수준 api를 사용한 다양한 시각화 기능을 제공한다.
- Matplotlib 시각화 함수
  - 라인 플롯(line plot)
  - 스캐터 플롯(scatter plot)
  - 컨투어 플롯(contour plot)
  - 서피스 플롯(surface plot)
  - 바 차트(bar chart)
  - 히스토그램(histogram)
  - 박스 플롯(box plot)
