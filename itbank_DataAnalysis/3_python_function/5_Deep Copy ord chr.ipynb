# 대입 연산
- 대입 연산은 '객체 복사'가 아니다.

## id()
- 참조하는 데이터의 위치값을 반환한다.
- 중복된 literal(immutable) 객체는 만든지 않는다.

- 값이 같아지면 데이터의 참조위치가 같아진다.

# Mutable과 Immutable 객체
image.png

- mutable 한 객체의 변수간 대입

- immutable 한 객체의 변수간 대입

# 얕은 복사(Shallow copy)

- mutable한 list에 immutable 한 데이터 대입은 서로 문제가 되지 않는다.

- 하지만 mutable 한 list에 mutable 한 list 를 넣으면 문제가 발생한다.

- 만일 a[0]을 재할당할 경우에는 문제가 발생하지 않는다.

- 하지만 재할당 없이 값만 변경하면 b 역시 변경된다.

# 깊은 복사(Deep copy)
- 깊은 복사는 내부에 객체들까지 모두 새롭게 Copy 되는 것을 말한다.

# ord()
- 문자를 입력하면 문자에 해당하는 UNICODE의 값을 출력해 준다.

# chr()
- 정수를 입력하면 정수에 해당하는 UNICODE의 문자를 출력해준다.
