  - 기본적으로 [ ]는 column을 추출

```python
import pandas as pd
import numpy as np

filepath = r'data/titanic.csv'

df = pd.read_csv(filepath)
df.head()
```

## 복수의 컬럼 선택하기
- list 형태로 컬럼을 넣으면

```python
#  list 형태로 인덱싱 하면
df[['Survived']]  # 결과가 DataFrame 이다! ★
```

```python
df[['Survived', 'Age', 'Name']]  # 명시한 컬럼의 순서대로 DataFrame  생성
```

# Row 선택하기

## DataFrame slicing
  - dataframe의 경우 기본적으로 [] 연산는 **column 선택**에 사용  `(인덱싱 용이 아니다!!)`
  - 하지만, slicing은 row 레벨로 지원

```python
df[:10]   # slicing 을 row level 로 동작한다!?
```

```python
df[7:10]
```

```python
df[-5:]  # 마지막 5개  tail()과 동일.
```

```python
df[::10] #  step
```

```python
df[::-1]  # 역순
```

## row 선택하기
  - Series의 경우 []로 row 선택이 가능하나, **DataFrame의 경우는 기본적으로 column을 선택하도록 설계**
  - **.loc[], .iloc[]**로 row 선택 가능
    - loc - 인덱스 자체를 사용
    - iloc - 0 based index로 사용
    - 이 두 함수는 ,를 사용하여 column 선택도 가능

#### loc[index]

##### iloc[0-base index]

## row, column 동시에 선택하기
loc[], iloc[] 속성을 이용할때, 콤마를 이용하여 둘다 명시 가능

##### iloc 사용하면 row 뿐 아니라 column 도 0-base 로 접근해야 한다
