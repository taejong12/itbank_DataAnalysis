# random 사용하여 array 생성
- random 서브 모듈 함수를 통해 ndarray를 생성할 수 있다.
- random 모듈은 무작위 데이터(난수)를 생성한다.
- 머신러닝에서는 weight 값을 무작위로 설정한 뒤 train 하면서 모델에 fit 하게 된다.

## np.random 서브 모듈
- rand : 0 부터 1 사이의 균일 분포
- randn : 가우시안 표준 정규 분포
- randint : 균일 분포의 정수 난수

## rand 함수
- 0 ~ 1 사이의 균일 분포로 랜덤한 ndarray를 생성한다.
```
np.random.rand(shape)
```

## randn 함수
- n: normal distribution(정규분포)
- 정규분포로 샘플링된 랜덤 ndarray를 생성한다.
- 단순히 0과 1 사이의 값이 아닌 정규 분포로 음수값도 갖게 된다.
- rand 함수는 uniform 한 분포에서 값을 가져오고 randn 은 normal distribution 에서 값을 가져온다.

## randint 함수
- 특정 정수 사이에서 랜덤하게 ndarray 생성한다.


## seed 함수
- 랜덤한 값을 동일하게 다시 생성하고자 할 때 사용한다.
- 후에 데이터 분석, 머신러닝, 딥러닝 등 무언가 무작위로 데이터를 생성하여 테스트 할 때 많이 사용한다.
- 이유는 똑같은 값이어야 모델의 성능이 좋아지는지 아닌지를 알 수 있기 때문이다.

## choice 함수
- 주어진 값 중 랜덤으로 값을 선택한다.
- 정수가 주어진 경우 np.arange()로 간주한다.

## 확률 분포에 따른 ndarray 생성하기
- uniform : 균등 분포 - 0과 1사이 균등 분포
- normal : 정규 분포
- 등 확률 분포는 다양하다.
