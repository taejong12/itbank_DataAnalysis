# ndarray shape 변경하기
- **차원변환**은 데이터분석 과 인공지능에서 매우 빈번하게 발생되는 작업이니만큼 자유자재로 변환할수 있어야 하고, 머리속으로 내가 변환하는 데이터의 구조가 그려지도록 익숙해져야 합니다

- 사용 예) 이미지 데이터 벡터화 - 이미지는 기본적으로 2차원 혹은 3차원(RGB)이나 트레이닝을 위해 1차원으로 변경하여 사용 됨

## ravel()
- 다차원 배열을 1차원으로 변경한다. 일반적으로 펼친다는 표현을 사용한다.
- 파라미터 : order
  - 'C' : row 우선 변경 (C Style)
  - 'F' : column 우선 변경 (Fortran Style)
- ravel() 은 np 안에 일반함수로도 있고, 혹은 ndarray의 멤버함수로서도 존재한다.

## flatten()
- 다차원 배열을 1차원으로 변경한다.
- ravel과의 차이점은 copy를 생성하여 변경하기 때문에 원본 데이터가 아닌 복사본을 반환하며 ndarray의 멤버 함수로만 제공된다.
- 파라미터 : order
  - 'C' : row 우선 변경
  - 'F' : column 우선 변경

- ravel()은 내부 데이터 참조를 유지한다.
- ravel 결과는 원본의 데이터 원소를 변경가능하다.
- 복사본으로 생성하지 않기 때문에 성능이 좋은 편이다.
- 반면 flatten()은 복사본을 생성하여 리턴한다.

- 3차원

## reshape 함수
- array의shape을 다른 차원으로 변경한다.
- 주의할 점은 reshape한 후의 결과의 전체 원소 갯수와 이전 갯수가 같아야 차원 변경이 가능하다.
- 사용 예) 이미지 데이터 벡터화 - 이미지는 기본적으로 2차원 혹은 3차원(RGB)이나 트레이닝을 위해 1차원으로 변경하여 사용된다.
- -1값을 주어 나머지 차원을 유추하게 할 수 있다.

- -1 값을 주고 행이나 열값을 입력하면 자동으로 차원 가능 여부를 확인 할 수 있다.

- shape 을 이용하여 직접 차원을 변경하여 원본의 차원도 변경할 수 있다.
