# NumPy
- NumPy는 수치적 연산을 위해 최적화된 모듈이다.
- 파이썬을 활용한 데이터 분석에서 가장 기본이 되는 모듈이다.
- NumPy 자체도 많이 쓰지만, 앞으로 배울 pandas, 머신러닝 등 수많은 분야에서 기본적으로 사용하는 모듈이다.
-  파이썬 개발자라면 꼭 익힐 필요 있다.

- 수학수치 연산  (numerical python) / 합, 평균, 행렬연산 등을 할 수 있다.
- 상황에 따라, pandas 안써도 numpy 만으로도 가능하다.

- 기본 타입 : ndarray  (n-dimensional array) 다차원 배열 표현으로 다양한 형태에 배열을 표현할 수 있다.

- 파이썬에 list 가 있는데 numpy 를 사용하는 이유
  - 성능 : 성능적으로 훨~씬 빠름.
  - 메모리 : 훨씩 적은 메모리사용
  - 제공함수풍부 : 선형대수, 통계관련 여러 수치함수 제공
  - 수많은 모듈에서 numpy 함수명 과 기능을 그대로 사용.

- ndarray 의 코어 부분은 'C언어'로 만들었다.

```
np.array()
```

## 자주 사용하는 ndarray 메서드
- shape : array 의 형태를 확인할 수 있다.
- size : array 의 데이터 수를 확인할 수 있다.
- ndim : array의 차원을 확인할 수 있다.
- len : array의 길이(행)를 확인할 수 있다.
- dtype : array의 데이터 타입을 확인 할 수 있다.


- ndarray 는 정수형이면 int, 실수형이면 float, 정수와 실수형이 섞이면 float 로 데이터 타입이 정해진다.

## NumPy 자료형
- int(8, 16, 32, 64) : 부호가 있는 정수
- uint(8, 16, 32, 64) : 부호가 없는 정수
- float(16, 32, 64, 128) : 실수
- complex(64, 128, 256) : 복소수
- bool : 논리(True, False)
- string : 문자열
- object : 파이썬 객체
- unicode : 유니코드

- 똑같은 값이라도 차원이 다를 수 있다.

## np.arange 함수로 array 생성
- 파이썬의 range 와 거의 동일하다.

```
np.arange(iterable, dtype=np.자료형)
```

## astype 으로 데이터형식 변경하기
- 타입 변환한 array 리턴으로 원본에는 변화가 없다.
```
변수 = 변수.astype(np.자료형)
```

## reshape() 으로 차원 형태 변경하기
- ndarray의 형태, 차원을 바꾸기 위해서 사용한다.
- 머신러닝, 데이터 프로세싱에서 매우 자주 사용된다.
```
변수.reshape(행, 열)
```

## np.ones, np.zeros 로 array 생성하기
- np.ones : 1로 이루어진 array 생성
- np.zeros : 0으로 이루어진 array 생성

## np.empty 로 array 생성하기
- 초기화가 안된 값들로 이루어진 array 생성한다.
- 초기화가 안되어 있기 때문에 크기가 크면 빠르게 array를 생성할 수 있다.
- 값이 0이나 1로 나오기도 하지만 랜덤이기 때문에 어떤 값이 나올 지 알 수 없다.


## np.full 로 array 생성하기
- 하나의 값으로 이루어진 array 를 생성한다.
```
np.full((shape), 값)
```

## np.eye 로 array 생성하기
- 단위 행렬을 생성한다.
- 2차원, 행과 열의 크기가 같다.
- 대각선 값이 1이고 나머지는 0으로 구성된다.
- 선형  대수 등 행렬곱에서 중요한 행렬이다.

## np.linspace 로 array 생성하기
- 그래프 등을 생성할 때 균등한 값으로 이루어진 array를 생성하고 싶을 때 사용한다.
```
np.linspace(시작값, 끝값, 데이터수)
```
